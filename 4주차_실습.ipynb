```python
import numpy as np
import pandas as pd

red = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/wine+quality/winequality-red.csv',sep = ';')    # sep로 행,열 구분
white = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/wine+quality/winequality-white.csv',sep = ';')
```

```python
# red, white 데이터에 데이터 구분 컬럼 생성

red["class"] = "0"
white["class"] = "1"

print(red.shape)
print(white.shape)

white.head()
```

```python
# red, white 데이터를 위아래로 합치기

wine = pd.concat([red,white])

print(wine.shape)
wine.head()
```

fixed acidity: 결합산

volatile acidity: 휘발성 산

citric acid: 시트르산

residual sugar: 잔당

chlorides: 염화물

free sulfur dioxide: 유리 이산화황

total sulfur dioxide: 이산화황

density: 비중

pH: 산도

sulphates: 황

alcohol: 알콜 도수

quality: 품질(0~10 사이의 값)

```python
data = wine[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']].to_numpy()
# 기존에 실습하던 wine_csv 파일과 다르게 더 많은 특성을 data에 numpy함
target = wine['class'].to_numpy()
```

```python
print(target)
```

```python
LABELS = ['Red_wine', 'White_wine']

import matplotlib.pyplot as plt
count_classes = pd.value_counts(wine['class'], sort = True)
count_classes.plot(kind = 'bar', rot=0)
plt.title("Wine distribution")
plt.xticks(range(2), LABELS)
plt.xlabel("Wine class")
plt.ylabel("Frequency")
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42)
```

```python
print(train_input.shape, test_input.shape)
```

```python
from sklearn.preprocessing import StandardScaler    # 전처리 과정

ss = StandardScaler()
ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_scaled, train_target)

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))   # 많은 특성을 훈련시키다보니 완벽에 가까운 결과를 도출함
```

# 결정 트리

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_scaled, train_target)

print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()      # 맨 위에 있는 것이 루트 노드, 맨 아래에 있는 것이 리프 노드
```

```python
plt.figure(figsize=(10,7))
plot_tree(dt, max_depth=1, filled=True, feature_names=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality'])   #max_depth를 1로 주면 루트 노드를 제외하고 하나의 노드만 확장하여 그림
plt.show()    # 양성이 커질수록 파란색, 음성이 커질수록 주황색
```

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(train_scaled, train_target)

print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
plt.figure(figsize=(20,15))
plot_tree(dt, filled=True, feature_names=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality'])
plt.show()
```

```python
print(dt.feature_importances_)  # 특성 중요도
```

total sulfur dioxide, chlorides을 중요하게 생각함

# 랜덤 포레스트

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))  #과대적합
```

# 엑스트라 트리

```python
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
et.fit(train_input, train_target)
print(et.feature_importances_)  #랜덤 포레스트와 마찬가지로 특성 중요도를 제공함
```

# 그레이디언트 부스팅

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb.fit(train_input, train_target)
print(gb.feature_importances_)
```

# 히스토그램 기반 그레이디언트 부스팅

```python
# 사이킷런 1.0 버전 아래에서는 다음 라인의 주석을 해제하고 실행하세요.
# from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(random_state=42)
scores = cross_validate(hgb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
from sklearn.inspection import permutation_importance

hgb.fit(train_input, train_target)
result = permutation_importance(hgb, train_input, train_target, n_repeats=10,
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
result = permutation_importance(hgb, test_input, test_target, n_repeats=10, #n_repeats 매개변수는 랜덤하게 섞을 횟수를 지정, 기본값은 5
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```
